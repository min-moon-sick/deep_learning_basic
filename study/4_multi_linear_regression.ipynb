```python
data = [[2, 0, 81], [4, 4, 93], [6, 2, 91], [8,3, 97]]

x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd

ax = plt.axes(projection = '3d')

ax.set_xlabel('time')
ax.set_ylabel('class')
ax.set_zlabel('score')

ax.scatter(x1, x2, y)
plt.show()
```

```python
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.03

epoch = 1001
```

```python
for i in range(epoch):
    
    y_pred = a1 * x1_data + a2 * x2_data + b
    error = y_data - y_pred
    
    a1_diff = -(2/len(x1_data)) * sum(x1_data * error)
    a2_diff = -(2/len(x2_data)) * sum(x2_data * error)
    b_diff = -(2/len(x1_data)) * sum(error)
    
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print("epoch : {}, 기울기1 : {}, 기울기2 : {}, 절편 : {}".format(i, a1_diff, a2_diff, b_diff))
        
y_pred = a1 * x1_data + a2 * x2_data + b
```
