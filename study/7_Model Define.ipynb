```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
dataset = pd.read_csv('../080228-master/080228-master/deeplearning/dataset/ThoraricSurgery.csv', header=None)

x = dataset.iloc[:, :17]
y = dataset.iloc[:, 17]
```

```python
x_arr = np.array(x)
y_arr = np.array(y)
```

```python
model = Sequential()
model.add(Dense(30, input_dim = 17, activation = 'relu'))
model.add(Dense(1, activation= 'sigmoid'))
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(x_arr, y_arr, epochs = 100, batch_size = 10)
```
